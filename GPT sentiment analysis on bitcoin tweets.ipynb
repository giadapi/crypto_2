```python
import openai

# Set your OpenAI API key
openai.api_key = "[YOUR API KEY]"

# Define a function to classify the sentiment of a given tweet using GPT-3
def classify_sentiment(tweet):
    prompt = "Perform sentiment analysis classification on this tweet " + tweet + "\nPositive, Negative, Neutral"
    response = openai.Completion.create(
        model="text-davinci-003",
        prompt = prompt,
        temperature=0,
        max_tokens=10,
        top_p=1.0,
        frequency_penalty=0.5,
        presence_penalty=0.0)
    
    sentiment = response.choices[0].text.strip()
    return sentiment

# Read in your CSV file of tweets
# import csv

# with open('tweets.csv', newline='') as csvfile:
#     reader = csv.reader(csvfile)
#     for row in reader:
#         tweet = row[0]
#         sentiment = classify_sentiment(tweet)
#         print(tweet + " - " + sentiment)
```
